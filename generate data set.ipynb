```python
import numpy as np
```

```python
m=1000
```

```python
x1 = np.random.random(m)*5 +1
```

```python
x1[0:5]
```

```python
x2 = np.random.random(m)*5 +1
```

```python
x3 = np.random.random(m)*5 +1
```

```python
y = 2.5*x1 + 3.6*x2 + 4.7*x3 + 10
#Adding some error in output
y=y+np.random.random(m)
```

```python
y[0:5]
```

```python
f = open("dataset.csv", "w")
```

```python
f.write("x1,x2,x3,y\n")
```

```python
for i in range(0,m):
    f.write("{},{},{},{}\n".format(x1[i],x2[i],x3[i],y[i]))
```

```python
f.close()
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x1,(3.6*x2+4.7*x3),y, s=50, c='r', marker='o', edgecolors='w')
ax.set_xlabel('X1')
ax.set_ylabel('3.6X2+4.7X3')
ax.set_zlabel('Y')
ax.text2D(0.05, 0.95, "Ploating X1, 3.6*X2+4.7*X3", transform=ax.transAxes)
plt.show()
```

```python
fig1=plt.figure()
ax1 = fig1.add_subplot(111, projection='3d')
ax1.scatter(x2,(2.5*x1+3.6*x2),y, s=50, c='r', marker='o', edgecolors='w')
ax1.set_xlabel('X2')
ax1.set_ylabel('2.5X1+4.7X3')
ax1.set_zlabel('Y')
ax1.text2D(0.05, 0.95, "Ploating X2, 2.5X1+4.7X3", transform=ax.transAxes)
plt.show()
```

```python
fig2=plt.figure()
ax2 = fig2.add_subplot(111, projection='3d')
ax2.scatter(x2,(2.5*x1+3.6*x2),y, s=50, c='r', marker='o', edgecolors='w')
ax2.set_xlabel('X3')
ax2.set_ylabel('2.5X1+3.6X2')
ax2.set_zlabel('Y')
ax2.text2D(0.05, 0.95, "Ploating X1, 2.5X1+3.6X2", transform=ax.transAxes)
plt.show()
```
